# PANDAS SERIES

## Importing Library

```python
import pandas as pd
import numpy as np
```

```python
country = ['India','Srilanka','Australia','USA','China']
```

```python
pd.Series(country)
```

## Custom Index

```python
marks = [67,68,64,33,27]
subjects = ['Maths','Science','Hindi','English','Social Science']
```

```python
pd.Series(marks)
```

```python
pd.Series(marks,index=subjects)
```

```python
# Setting a name: 
pd.Series(marks,index=subjects,name="Mayank ke marks")
```

## Series from Dictionary

```python
marks = {
    'maths' : 67,
    'english' : 89,
    'science' : 76,
    'Hindi' : 90
}
```

```python
marks_series = pd.Series(marks,name=)
```
