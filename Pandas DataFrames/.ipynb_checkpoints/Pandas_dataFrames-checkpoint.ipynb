```python
olymp_new = olymp.copy()
olymp_new
```

```python
olymp_new['First Name'] = olymp_new['name'].str.split(' ').str[0]
```

```python
olymp_new = olymp_new[['First Name','born_country','height_cm','weight_kg','died_date','born_date']]
```

```python
olymp_new
```

```python
olymp_new[['First Name','born_date']]
```

```python
olymp_new['born_datetime'] = pd.to_datetime(olymp_new['born_date'])
```

```python
olymp_new['born_Year'] = olymp_new['born_datetime'].dt.year
```

```python
olymp_new[['First Name','born_Year']]
```

```python
olymp_new.to_csv('olymp_new.csv')
```

```python
def categorize_athelete(row):
    if row['height_cm'] < 75 and row['weight_kg'] < 70:
        return 'Lightweight'
    elif row['height_cm'] < 185 or row['weight_kg'] <= 80:
        return 'Middleweight'
    else:
        return 'Heavyweight'

olymp_new['Category'] = olymp_new.apply(categorize_athelete, axis=1)
    
```
